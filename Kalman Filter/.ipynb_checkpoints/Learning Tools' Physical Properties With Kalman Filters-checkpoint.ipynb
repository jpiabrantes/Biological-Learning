# Learning Tools' Physical Properties With Kalman Filters

## The Problem

Imagine we have the robot arm represented on the image below:

We figured out what are the dynamics of this robot arm so we can use a Kalman Filter in order to use all our sensors to give us a better estimate of our arm position and velocity. However, if our robot arm grabs a tool the dynamics of the system change and our Kalman Filter stops working; e.g. we give an input to the robot arm to extend its elbow, according to our model we would see certain translation of the arm but since the arm is now holding a heavy tool it will move much less. 

Well we need a more <strong>flexible</strong> model!
If we attach some rigid object to the robot's arm, it will only change a few constants in our model. More exactly, it will change the center of mass, the mass and the inertial matrix of the robots forearm. We could add this constants $c$ to our state and have our Kalman Filter make estimations about them too:
$$x=\begin{pmatrix}
q\\
\dot{q}\\
c\end{pmatrix}$$

This way our robot will be able to use the Kalman Filter to make good estimations independently of the tool it's using. Moreover the robot will be able to learn the physical properties of each tool which may be very important when choosing the right tool for the right task.

## The Simplier Problem

In this notebook we will simulate a simplier problem that describes the learning of tool's physical properties using Kalman Filters.
Our robot arm in this case is a simple spring fixed on a wall and attached to a mass.

$$F=-k(q-u)$$

$q$ is the position of the mass, and $u$ is the robot input to move its arm. This "arm" can hold different objects that change the mass of the system, so the mass will also be part of our state.

$$x=\begin{pmatrix}q\\ \dot{q} \\ a\end{pmatrix}$$

Where $a=\frac{1}{m}$.
Our state update equations are:

$\begin{align}
q_n &= q_{n-1}+\dot{q}_{n-1}\Delta t \\
\dot{q}_n &= \dot{q}_{n-1}-k a_{n-1} \Delta t(q_{n-1}-u_{n-1}) \\
a_n &= a_{n_1}
\end{align}\\$

The function that updates our state is not linear so will need to use the Extended Kalman Filter.

$$x_n=f(x_{n-1},u_{n-1})+N(0,\Sigma_x)$$

The Jacobian of $f$ is:

$$F_{n-1}=\begin{pmatrix}
1 & \Delta t & 0\\
-ka\Delta t & 1 & k \Delta t (q_{n-1}-u_{n-1}) \\
0 & 0 & 1\end{pmatrix}$$

Our robot has a noisy sensor that reads the arm position:

$\begin{align}
y_n &= Cx+N(0,\sigma_y)\\
y_n &= \begin{pmatrix}1 & 0 & 0\end{pmatrix}x+N(0,\sigma_y)
\end{align}$

To get started with our kalman filter we define our priors $x^{(1|0)}$ and $P^{(1|0)}$ using some intuition.
When we get an observation we update our knowledge of the state:

$\begin{align}
k^n &= \frac{P^{(n|n-1)}C^T}{CP^{(n|n-1)}C^T+\sigma_y^2} \\
x^{(n|n)} &= x^{(n|n-1)}+k^n(y^n-Cx^{(n|n-1)}) \\
P^{n|n} &= (I-k^nC)P^{(n|n-1)}
\end{align}$

Finally we project our estimation to the next trial:

$$x_n=f(x_{n-1},u_{n-1})+N(0,\Sigma_x)$$

$$P^{(n+1|n)}=F_{n-1}P^{(n|n)}F_{n-1}^T+\Sigma_x$$



### Simulation:

```python
#preparations
%matplotlib inline
from IPython.core.pylabtools import figsize
from numpy import *
from matplotlib import pyplot as plt
import matplotlib,json

s = json.load( open("../styles/bmh_matplotlibrc.json") )
matplotlib.rcParams.update(s)
```

```python
#Global Variables
k = 100 #Newton/meter
C = array((1,0,0)).reshape(1, -1)
sigmay = 0.5

#covariance of the error measured between the rk4 simulation and f function
Sigmax=array([[  2.80022362e-07,  -5.60055740e-04,   0.00000000e+00],
       [ -5.60055740e-04,   1.12021199e+00,   0.00000000e+00],
       [  0.00000000e+00,   0.00000000e+00,   0.00000000e+00]])

#helper functions
def control(t):
    '''
    In this function the robot should decide based on his estimation of the state
    what is the next input u. For this simple example the input is constant.
    '''
    return 5 

def f(u,dt,x):
    '''
    State transition function.
    '''
    q,qdot,a=[float(i) for i in x]
    qnext=q+qdot*dt
    qdotnext=qdot-k*a*dt*(q-u)
    anext = a
    return array((qnext,qdotnext,anext)).reshape(-1, 1)

def rk4(t, h, x, f):
    #function used to simulate the real state position
    k1 = h * f(t, x)
    k2 = h * f(t + 0.5*h, x + 0.5*k1)
    k3 = h * f(t + 0.5*h, x + 0.5*k2)
    k4 = h * f(t + h, x + k3)
    return x + (k1 + 2*(k2 + k3) + k4)/6.0
    
def q_derivative(t,x):
    #function used to simulate the real state position
    q,qdot,a = [float(i) for i in x]
    qdotdot = -k*a*(q-control(t))
    adot=0
    return array((qdot,qdotdot,adot)).reshape(-1, 1)

def simulate_arm(xprior,Pprior):
    #the real state of the arm, we will keep a list (to plot afterwards) of the state for each iteration
    x = [array((20,0,3)).reshape(-1, 1)]
    #the estimated state of the arm and the estimated variance of the state, also as a list
    xpredict = [xprior]
    Ppredict = [Pprior]

    multi = 10 #every multi iteration we make a measurement
    maxiters = 10000
    for i in xrange(1,maxiters):
        dt=0.001
        t = i*dt
        u = control(t)
        x.append(rk4(t,dt,x[-1],q_derivative))
        if (not i%multi):
            dt *= multi
            S = dot(dot(C,Pprior),C.T)+sigmay**2
            K = dot(dot(Pprior,C.T),linalg.inv(S))
            y = random.normal(x[-1][0],sigmay)
            xpost = xprior + dot(K,(y-dot(C,xprior)))
            Ppost = dot((eye(3)-dot(K,C)),Pprior)
            xprior = f(u,dt,xpost)
            q,qdot,a=[float(i) for i in xpost]
            F = array(((1,dt,0),(-k*a*dt,1,-k*dt*(q-u)),(0,0,1)))
            Pprior = dot(dot(F,Ppost),F.T)+Sigmax

            xpredict.append(xpost)
            Ppredict.append(Ppost)

    #plot results
    figsize(30,15)
    state_names = ('position','speed','mass')
    for j in xrange(3):
        plt.subplot(1, 3,j+1) 
        t=[i*dt for i in xrange(0,maxiters,multi)]
        plt.plot([i*dt for i in xrange(0,maxiters,1)],[i[j] for i in x], label="true %s" %state_names[j] )
        plt.plot([i*dt for i in xrange(0,maxiters,multi)],[i[j] for i in xpredict], label="predicted %s" %state_names[j] )
        plt.xlabel('Time (s)')
        plt.legend()
```

```python
xprior = array((5,-15,10)).reshape(-1, 1) #the true state is array((20,0,3))
Pprior = array(((100**2,100,100),(100,50**2,100),(100,100,50**2))) #quite large
simulate_arm(xprior,Pprior)
```

We successful learned the "tool" mass and estimated correctly the arms position and speed along time.
