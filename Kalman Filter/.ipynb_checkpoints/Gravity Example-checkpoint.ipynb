# Kalman Filter - Gravity Example

## Problem description:

We have a object free falling from an initial height $q_0$ with an initial speed $\dot{q}_0$. We aren't sure where the object is but we have access to noisy measurements of its height from time to time.

Let $x$ be the state of our system.
$$x=\begin{pmatrix}
q\\
\dot{q}
\end{pmatrix}$$

Our generative functions are the following:

$$x_n=Ax_{n-1}+Bu_n+N(0,\Sigma_x)$$

$$x_n=\begin{pmatrix}
1 & \Delta t \\
0 & 1\end{pmatrix}
x_{n-1} + \begin{pmatrix}
0.5 \Delta t^2\\
\Delta t \end{pmatrix}
u_n+N(0,\Sigma_x)$$

$N(0,\Sigma_x)$ represents the uncertainty about how the state changes from trial to trial. Lets assume this uncertainty is zero, so our state changes exactly according to the linear equation above.
In terms of our measurement of the object height:

$$y_n=Cx_n+N(0,\sigma_y)$$

$$y_n=\begin{pmatrix}
1 & 0 \end{pmatrix}x_n+N(0,\sigma_y)$$

To get started with our kalman filter we define our priors $x^{(1|0)}$ and $P^{(1|0)}$ using some intuition.

When we get an observation we update our knowledge of the state:

$$k^n=\frac{P^{(n|n-1)}C^T}{CP^{(n|n-1)}C^T+\sigma_y^2}$$

$$x^{(n|n)}=x^{(n|n-1)}+k^n(y^n-Cx^{(n|n-1)})$$

$$P^{n|n}=(I-k^nC)P^{(n|n-1)}$$

Finally we project our estimation to the next trial:

$$x^{(n+1|n)}=Ax^{(n|n)}+Bu_n$$

$$P^{(n+1|n)}=AP^{(n|n)}A^T$$

```python
#preparations
%matplotlib inline
from IPython.core.pylabtools import figsize
from numpy import *
from matplotlib import pyplot as plt
import matplotlib,json

s = json.load( open("styles/bmh_matplotlibrc.json") )
matplotlib.rcParams.update(s)
```

```python
def gravity_kalman(xprior,Pprior):
    #variables for the generative model
    dt=0.1 
    A = array(((1,dt),(0,1)))
    B = array((0.5*dt*dt,dt)).reshape(-1, 1) 
    C = array((1,0)).reshape(1, -1)
    u=-9.8
    sigmay = 500 #quite big
    
    #the real state of the object, we will keep a list of the state for each iteration
    q=[25000] #initial height
    qdot=[0]  #initial speed
    
    #the predicted height of the object and the estimated standart deviation of the height, also as a list
    qpredict = [xprior[0,0]]
    qsigma = [sqrt(Pprior[0,0])]
    
    maxiters = 506
    for i in xrange(1,maxiters):
        t=i*dt
        q.append(q[0]+qdot[0]*t+0.5*u*t*t)
        qdot.append(qdot[0]+u*t)
        S = dot(dot(C,Pprior),C.T)+sigmay**2
        k = dot(dot(Pprior,C.T),linalg.inv(S))
        y = random.normal(q[-1],sigmay)
        xpost = xprior + dot(k,(y-dot(C,xprior)))
        Ppost = dot((eye(2)-dot(k,C)),Pprior)
        
        xprior = dot(A,xpost)+dot(B,u)
        Pprior = dot(dot(A,Ppost),A.T)
        qpredict.append(xpost[0,0])
        qsigma.append(sqrt(Ppost[0,0]))
    t=[i*dt for i in xrange(maxiters)]
    figsize(8, 5)
    plt.errorbar(t, qpredict,yerr=qsigma,label="height prediction",color="red")
    plt.plot(t, q,label="true height",color="blue")
    plt.xlabel("time (s)")
    plt.ylabel("height (m)")
    plt.ylim([0,50000])
    plt.xlim([0,t[-1]])
    plt.legend()
```

```python
xprior = array((50000,50)).reshape(-1, 1) #much higher than the true height 25000
Pprior = array(((20000**2,200**2),(200**2,100**2))) #huge.. we're definitely unsure about our prior
gravity_kalman(xprior,Pprior)
```

Our prediction converged quickly to the true value!

```python
xprior = array((50000,50)).reshape(-1, 1) #same as before
Pprior = array(((100**2,5**2),(5**2,2**2))) #much smaller! we're pretty sure about our prior!
gravity_kalman(xprior,Pprior)
```

Our bias has a much bigger importance now. We're very sure about it so we need lots of evidence to change our mind.
